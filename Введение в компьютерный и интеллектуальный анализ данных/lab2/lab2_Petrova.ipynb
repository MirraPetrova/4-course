Задание 3: Проверка актуальности данных.

```python
from datetime import datetime

data_3 = pd.read_csv('covid19_tweets.csv')
data_3['date'] = pd.to_datetime(data_3['date'], errors='coerce')
date_now = datetime.now()
threshold = pd.Timedelta(days = 1900)
data_3['is_recent'] = (date_now - data_3['date']) <= threshold

recent_tweets = data_3[data_3['is_recent']]
outdated_tweets = data_3[~data_3['is_recent']]

print(f"General: {len(data_3)}")
print(f"Actual: {len(recent_tweets)}")
print(f"Old: {len(outdated_tweets)}")
```

Задание 4: Проверка полноты данных.

```python
data_4 = pd.read_csv('Global Cancer data by Country (Females).csv')
countries_in_data = data_4['Women'].unique()

some_countries = ['Bosnia & Herzegovina', 'San Marino', 'Finland', 'Israel', 'Singapore', 'Argentina', 'United States', 'China', 'India', 'Brazil', 'Russia', 'Japan', 'Germany', 'France', 'United Kingdom', 'Italy']

missing = set(some_countries) - set(countries_in_data)

print(f"Some countries at list: {len(some_countries)}")
print(f"Countries in data: {len(countries_in_data)}")
print(f"No data: {len(missing)}")
print("Missing countries:", list(missing))
```
