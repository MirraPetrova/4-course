```python
import numpy as np

def mu_1(t):
    return -t*np.exp(-t)

def f(x,t):
    return -(x+pow(t,2))*np.exp(-x*t)

N = 20 # число узлов по пространству
N1 = 2000 # число временных шагов
h = 1/N
tau = 1/N1
sigma = 0.25
array = []
temp = []
for i in range(0, N+1):
    temp.append(1)
array.append(temp)
k=0
for j in range(0,N1):
    A = []
    B = []
    C = []
    F = []
    A.append(0)
    B.append(0)
    C.append(0)
    F.append(0)
    for i in range(1, N):
        A.append(sigma*tau/pow(h,2))
        B.append(sigma*tau/pow(h,2))
        C.append(1+2*sigma*tau/pow(h,2))
        F.append((1-sigma)*tau/pow(h,2)*(array[j][i-1]-2*array[j][i]+array[j][i+1])+array[j][i]+tau*f(h*i,tau*j))

    ksi_1 = 0
    ksi_2 = 1/(1+pow(h,2)/2/tau)
    nu_1 = 1
    nu_2 = (h*mu_1(tau*(j+1))+pow(h,2)/2/tau*array[j][N]+pow(h,2)/2*f(h*N,tau*(j+1)))/(1+pow(h,2)/2/tau)
    if k == 0:
        k+=1
        print("Решение:")
        print('A: '+str(A[1:]))
        print('B: ' + str(B[1:]))
        print('C: ' + str(C[1:]))
        print('ksi_1: ' + str(ksi_1 ))
        print('ksi_2: ' + str(ksi_2))
    alpha = []
    betta = []
    alpha.append(0)
    betta.append(0)
    y = np.zeros(N + 1)
    alpha.append(ksi_1)
    betta.append(nu_1)
    for i in range(1, N):
        alpha.append(B[i] / (C[i] - A[i] * alpha[i]))
        betta.append((F[i] + betta[i] * A[i]) / (C[i] - A[i] * alpha[i]))

    y[N] = (nu_2 + ksi_2 * betta[N]) / (1 - alpha[N] * ksi_2)
    for i in range(N - 1, -1, -1):
        y[i] = alpha[i + 1] * y[i + 1] + betta[i + 1]

    array.append(list(y))

fout = open("Output.txt", 'w', encoding="utf-8")
array1 = array.copy()


array1 = list(reversed(array1.copy()))
for i in range(int(N1/200)+1):
    for j in range(int(N/2)+1):
        fout.write(str(round(array1[i*200][j*2],4))+' ')
    fout.write('\n')
fout.close()



N = 20
N1 = 4000
h = 1/N
tau = 1/N1
sigma = 0.25
array = []
temp = []
for i in range(0, N+1):
    temp.append(1)
array.append(temp)
k = 0
for j in range(0, N1):
    A = []
    B = []
    C = []
    F = []
    A.append(0)
    B.append(0)
    C.append(0)
    F.append(0)
    for i in range(1, N):
        A.append(sigma*tau/pow(h,2))
        B.append(sigma*tau/pow(h,2))
        C.append(1+2*sigma*tau/pow(h,2))
        F.append((1-sigma)*tau/pow(h,2)*(array[j][i-1]-2*array[j][i]+array[j][i+1])+array[j][i]+tau*f(h*i,tau*j))
    ksi_1 = 0
    ksi_2 = 1/(1+pow(h,2)/2/tau)
    nu_1 = 1
    nu_2 = (h*mu_1(tau*(j+1))+pow(h,2)/2/tau*array[j][N]+pow(h,2)/2*f(h*N,tau*(j+1)))/(1+pow(h,2)/2/tau)
    if k == 0:
        print("Точное решение:")
        k+=1
        print('A: '+str(A[1:]))
        print('B: ' + str(B[1:]))
        print('C: ' + str(C[1:]))
        print('ksi_1: ' + str(ksi_1))
        print('ksi_2: ' + str(ksi_2))
    alpha = []
    betta = []
    alpha.append(0)
    betta.append(0)
    y = np.zeros(N + 1)
    alpha.append(ksi_1)
    betta.append(nu_1)
    for i in range(1, N):
        alpha.append(B[i] / (C[i] - A[i] * alpha[i]))
        betta.append((F[i] + betta[i] * A[i]) / (C[i] - A[i] * alpha[i]))

    y[N] = (nu_2 + ksi_2 * betta[N]) / (1 - alpha[N] * ksi_2)
    for i in range(N - 1, -1, -1):
        y[i] = alpha[i + 1] * y[i + 1] + betta[i + 1]

    array.append(list(y))

fout = open("Output1.txt", 'w', encoding="utf-8")
array2 = []
for i in range(0, len(array1)):
        array2.append(array[i*2])


array2 = list(reversed(array2))
for i in range(int(N1/400)+1):
    for j in range(int(N/2)+1):
        fout.write(str(round(array2[i*200][j*2],4))+' ')
    fout.write('\n')
fout.close()




array_nevazka = []
fout3 = open('Nevazka.txt', 'w', encoding='utf-8')
for i in range(int(N1/400)+1):
    temp = []
    for j in range(int(N/2)+1):

        temp.append(abs(array1[i*200][j*2]-array2[i*200][j*2]))

    array_nevazka.append(temp)


for i in array_nevazka:
    for j in i:
        fout3.write(str(j)+' ')
    fout3.write('\n')
fout3.close()
```
